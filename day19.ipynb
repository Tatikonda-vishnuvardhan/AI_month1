```python
#Data
dev_y = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
py_dev_y = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
js_dev_y = [37810, 43515, 46823, 49293, 53437,
            56373, 62375, 66674, 68745, 68746, 74583]
```

```python
plt.plot(age_x,py_dev_y,label = "Python developers")
plt.title("Median Salary by Age for Python Developers")
plt.xlabel("Age(yrs)")
plt.ylabel("Salary Median(USD)")
plt.legend()
plt.grid(False)
plt.show()
```

***Combining above two plots***

```python
plt.plot(age_x,dev_y,"rs--")#colour,marker,line_style
plt.plot(age_x,py_dev_y,"g*:")
plt.title("Median Salary by Age")
plt.xlabel("Age(yrs)")
plt.ylabel("Salary Median(USD)")
plt.legend(["All Developers","Python Developers"])
plt.show()
```

```python
plt.plot(age_x,js_dev_y,"ko-",label = "Java_Script")
plt.plot(age_x,py_dev_y,color = "g",marker = "*",linestyle = ":",label = "Python Developers")
plt.plot(age_x,dev_y,"y.--",label = "All Developers")
plt.title("Median Salary by Age")
plt.xlabel("Age(yrs)")
plt.ylabel("Salary Median(USD)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

**From CSV File**

```python
import pandas as pd
from collections import Counter
```

```python
df = pd.read_csv("day19_data1.csv")
df
```

```python
id = df["Responder_id"]
lan = df["LanguagesWorkedWith"]
lan_counter = Counter()
```

```python
for item in lan:
    lan_counter.update(item.split(";"))
lan_counter
```

```python
lan_counter.most_common(15)
languages = []
numbers = []
for item in lan_counter.most_common(15):
    languages.append(item[0])
    numbers.append(item[1])
plt.style.use('fivethirtyeight')
plt.bar(languages,numbers)
plt.title("Language Popularity")
plt.xlabel("Proggaming Languages")
plt.ylabel("Popularity in Numbers")
plt.tight_layout()
plt.show()
```

```python
plt.style.use('fivethirtyeight')
plt.barh(languages,numbers)
plt.title("Popular Languages ")
#plt.ylabel("Proggaming Languages")
plt.xlabel("Popularity in Numbers")
plt.tight_layout()
plt.show()
```

```python
languages.reverse()
numbers.reverse()
plt.style.use('fivethirtyeight')
plt.barh(languages,numbers)
plt.title("Popular Languages ")
plt.xlabel("Popularity in Numbers")
plt.tight_layout()
plt.show()
```

**Scatter Plots**

```python
df3 = pd.read_csv("day19_data3.csv")
views = df3["view_count"]
likes = df3["likes"]
plt.scatter(views,likes,linewidths=1,edgecolors="black",alpha=0.5)
plt.xscale("log")
plt.yscale("log")
plt.tight_layout()
plt.show()
```

```python
from matplotlib.colors import Colormap


ratio = df3["ratio"]
plt.scatter(views,likes,c=ratio,cmap= "summer",linewidths=1,edgecolors="black",alpha=0.5)
cbar = plt.colorbar()
cbar.set_label("Ratio(like/dislike)")

plt.xscale("log")
plt.yscale("log")
plt.tight_layout()
plt.show()
```
