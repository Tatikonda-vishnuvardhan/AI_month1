**Group By**

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Name":["Vishnu","Mukesh","Nithin",],
    "Age" : [23,27,32],
    "Occupation" : ["AI","Data Science","Web Developer"]
})
df
```

```python
df.loc[3] = ["Arya",19,"Student"]
df
```

```python
df.loc[0:2,"Category"] = "programmer"
df = df.rename(columns={"Occupation": "Job"})
print(df)


df
```

```python
df.info()
```

```python
df.groupby("Category").agg({"Age": ["mean","median"]})
```

```python
df = df.sort_values("Age",ascending= False)
df
```

```python
df.reset_index(drop=True)
```

***Mini Daily Task***

```python
data = {
    "Name": ["Vishnu", "Sai", "Mukesh", "Anita", "John", "Sara", "David", "Priya"],
    "Math": [85, 90, 78, 88, 92, 70, 65, 80],
    "Science": [88, 76, 85, 90, 95, 60, 75, 82],
    "English": [80, 85, 75, 89, 91, 65, 70, 84]
}
```

```python
stu = pd.DataFrame(data)
stu
```

```python
stu.count()
```

```python
stu["Class"] = ["A","B","A","B","A","B","A","B"]
stu
```

```python
stu.groupby("Class").agg({"Math": ["mean","max","min"],"Science": ["mean","max","min"],"English": ["mean","max","min"]})
```
