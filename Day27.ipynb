gradient descent

```python
import matplotlib.pyplot as plt
import numpy as np
#%matplotlib widget

# Function: f(x) = x^2
def f(x):
    return x**2

# Derivative: f'(x) = 2x
def df(x):
    return 2*x

# Gradient Descent
x = 10   # starting point
learning_rate = 0.1
history = [x]

for _ in range(20):  # 20 iterations
    x = x - learning_rate * df(x)
    history.append(x)

# Plot
xs = np.linspace(-10, 10, 100)
ys = f(xs)

plt.plot(xs, ys, label="f(x)=x^2")
plt.scatter(history, [f(x) for x in history], color="red")
plt.plot(history, [f(x) for x in history], color="red", linestyle="--", label="Descent Path")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gradient Descent Visualization")
plt.show()
```
