**Seaborn**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter magics
%matplotlib widget
```

```python
print(sns.get_dataset_names())
```

```python
car_df = sns.load_dataset("car_crashes")
car_df
```

**Distribution Plot** /basically histogram

```python
from statistics import kde


bin = [0,5,10,15,20,25,30]
sns.displot(car_df["alcohol"],bins = bin,kde = True)
```

**Joint Plot**/basically scatter plot

```python
sns.jointplot(x = car_df["alcohol"],y = car_df["not_distracted"],kind="hex")
plt.show()
```

**KDE Plot**

```python


plt.figure()
sns.kdeplot(car_df["ins_losses"], fill=True)
```

**Pair Plot**

```python
plt.figure()
sns.pairplot(car_df)
```

**Rug Plot**

```python
plt.figure()
sns.rugplot(car_df["alcohol"])
```

**Bar Plot**

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
plt.figure()
sns.barplot(x=tips_df["sex"],y=tips_df["tip"])
plt.show()
```

**Box Plot**

```python
plt.figure()
sns.boxplot(x=tips_df["day"],y=tips_df["total_bill"],hue=tips_df["sex"])
plt.show()
```

**Heat Map**

```python
car_df
```

```python
plt.figure()
matrx = car_df[["total","speeding",	"alcohol",	"not_distracted",	"no_previous",	"ins_premium",	"ins_losses"]].corr()
sns.heatmap(matrx,annot=True)
plt.tight_layout()
plt.show()
```

**Cluster Map**

```python
plt.figure()
sns.clustermap(car_df[["total","speeding",	"alcohol",	"not_distracted",	"no_previous",	"ins_premium",	"ins_losses"]],cmap="Blues")
plt.show()
```

```python

sns.set_context("notebook",font_scale=1.5)
sns.lmplot(x="tip",y="total_bill",data=tips_df,hue="sex",markers=["o","s"],scatter_kws={'s' : 100,'linewidths' : 1,'edgecolor':"black"})
plt.tight_layout()
plt.show()
```
