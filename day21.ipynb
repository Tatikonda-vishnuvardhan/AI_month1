```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("day17_StudentsPerformance.csv")

# 1. Basic Analysis
print("First 5 rows:\n", df.head())
print("\nColumn Info:\n", df.info())
print("\nMissing Values:\n", df.isnull())

# 2. Statistics
print("\nStatistics:\n", df.describe())

# 3. Filtering: Students with math score > 80
high_math = df[df["math score"] > 80]
print("\nStudents with math score > 80:\n", high_math.head())

# 4. Grouping: Average scores by gender
avg_scores = df.groupby("gender").agg({
    "math score": "mean",
    "reading score": "mean",
    "writing score": "mean"
})

print("\nAverage Scores by Gender:\n", avg_scores)

# 5. Visualizations
# Histogram
plt.hist(df["math score"], bins=10, edgecolor="black",linewidth = 5)
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.show()

# Bar Plot
sns.barplot(x="gender", y="math score", data=df,estimator=np.mean,edgecolor="black",linewidth = 5)
plt.title("Average Math Score by Gender")
plt.show()

# Heatmap
corr = df[["math score","reading score","writing score"]].corr()
sns.heatmap(corr, annot=True, cmap="Blues")
plt.title("Correlation Between Scores")
plt.show()
```
